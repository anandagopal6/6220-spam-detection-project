# Extract emails for spamword cloud

## Get word cloud for overall spam emails

```python
dataset_path = "/content/drive/MyDrive/6220 Project/SpamDatasets/processed_data/fully_combined_data.csv"
df = pd.read_csv(dataset_path, names=['Content', 'Label'])
df
```

```python
spam = df.loc[df['Label'] == 1]
spam
```

```python
spam_string = ""
print(type(spam_string))
for row in spam.iterrows():
    spam_string = spam_string + row[1][0]
```

```python
print(len(spam_string.split()))
print(len(spam_string))
```

```python
stopwords = ["width", "height", "td", "tr", "font", "face", "img", "src", "size", "br", "bg", "http", "www", "arial", "helvetica", "org", "debian", "gif", "sans", "sans serif", "com", "cellpadding", "cellspacing", "align", "center", "table", "border", "nbsp"]
cloud = wordcloud.WordCloud(stopwords = stopwords, max_font_size=50, background_color="white").generate(spam_string)
```

```python
plt.figure()
plt.imshow(cloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
cloud.to_file("/content/drive/MyDrive/6220 Project/spam_cloud.png")
```
